```python
colors=['r','k','b','orange']
plt.bar('Name','Score',color=colors,data=df_data)
plt.title('Bar Chart')
plt.xlabel('Names')
plt.ylabel('Score')
plt.xticks(rotation = 90)
plt.show()
```

```python
colors=['r','k','b','orange']
plt.bar('Name','Episodes',color=colors,data=df_data)
plt.title('Bar Chart')
plt.xlabel('Names')
plt.ylabel('Episodes')
plt.xticks(rotation = 90)
plt.show()
```

```python
colors=['r','k','b','orange']
plt.barh('Name','Score',color=colors,data=df_data)
plt.title('Horizontal Bar Chart')
plt.xlabel('Score')
plt.ylabel('Name')
plt.show()
```

```python
colors=['r','k','b','orange']
plt.barh('Name','Episodes',color=colors,data=df_data)
plt.title('Horizontal Bar Chart')
plt.xlabel('Episodes')
plt.ylabel('Name')
plt.show()
```

```python
plt.pie('Score',labels='Name',data=df_data)
plt.title('Anime Classification')
plt.show()
```

```python
plt.pie('Episodes',labels='Name',data=df_data)
plt.title('Anime Classification')
plt.show()
```
